### Read the `air_quality.csv` data from the `Resources` folder into a Pandas DataFrame.

```python
# Read the 12 CSV file for each station from the Resources folder into a Pandas DataFrame. 
# Zipped file 'raw_12_data_files.7z' placed in the GitHub Resources folder

df1 = pd.read_csv(Path("Resources/PRSA_Data_Dongsi_20130301-20170228.csv"))
df2 = pd.read_csv(Path("Resources/PRSA_Data_Guanyuan_20130301-20170228.csv"))
df3 = pd.read_csv(Path("Resources/PRSA_Data_Gucheng_20130301-20170228.csv"))
df4 = pd.read_csv(Path("Resources/PRSA_Data_Huairou_20130301-20170228.csv"))
df5 = pd.read_csv(Path("Resources/PRSA_Data_Nongzhanguan_20130301-20170228.csv"))
df6 = pd.read_csv(Path("Resources/PRSA_Data_Shunyi_20130301-20170228.csv"))
df7 = pd.read_csv(Path("Resources/PRSA_Data_Tiantan_20130301-20170228.csv"))
df8 = pd.read_csv(Path("Resources/PRSA_Data_Wanliu_20130301-20170228.csv"))
df9 = pd.read_csv(Path("Resources/PRSA_Data_Wanshouxigong_20130301-20170228.csv"))
df10 = pd.read_csv(Path("Resources/PRSA_Data_Aotizhongxin_20130301-20170228.csv"))
df11 = pd.read_csv(Path("Resources/PRSA_Data_Changping_20130301-20170228.csv"))
df12 = pd.read_csv(Path("Resources/PRSA_Data_Dingling_20130301-20170228.csv"))
                                                                     
# Review the DataFrame
display(df1.head())
display(df1.tail())
```

```python
# count of each column 
df1.isna().sum()
```

```python
# make a list of the df's
list_df = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12]

# for loop to check number of NaN issues in all dataframes
na_info = {}

# for i in list_df:
#     i.isna().sum()

for i in list_df:
    na_info = i.isna().sum()
    print(na_info)

```

```python
# dropNaN and write out df1_drop
df1_drop = df1.dropna()
df1_drop.isna().sum()
# display(df1_drop.count())  shows that we dropped 5000 rows or about 15% of the data
```

```python
# for loop to dropNaN from all 12 data files and write out df1_drop
list_df_drop = []

for i in list_df:
    i2 = i.dropna()
    list_df_drop.append(i2)
```

```python
# drop NAN from all 12 data files
df11_drop = list_df_drop[10]
df11_drop.head()
```

```python
# concatonate the 12 data fames that contain the original data
data = pd.concat(list_df)

# concatonate data frames with dropping NA 
data_tableau = pd.concat(list_df_drop)

# Review the DataFrame
display(data.head())
display(data.tail())
display(data.count())

# count total rows. Identified approximately 5% of data for NO2, CO, and O4 is NAN
display(data_tableau.count()) 

print(f"the number of rows in df DataFrame is {len(data)} after merging the original csv files.")

# # df data_tableau shows about 10% of rows were dropped
print(f"the number of rows in data_tableau DataFrame is {len(data_tableau)} after dropping numeric NA's")
```

```python
# drop the 'No' column from data. 'No' Not needed
data = data.drop(columns=['No']) 
```

```python
# reset index
data= data.reset_index() 
data.head()
```
