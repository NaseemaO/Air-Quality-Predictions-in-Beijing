### Read the `air_quality.csv` data from the `Resources` folder into a Pandas DataFrame.

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame

df1 = pd.read_csv(Path("Resources/PRSA_Data_Dongsi_20130301-20170228.csv"))
df2 = pd.read_csv(Path("Resources/PRSA_Data_Guanyuan_20130301-20170228.csv"))
df3 = pd.read_csv(Path("Resources/PRSA_Data_Gucheng_20130301-20170228.csv"))
df4 = pd.read_csv(Path("Resources/PRSA_Data_Huairou_20130301-20170228.csv"))
df5 = pd.read_csv(Path("Resources/PRSA_Data_Nongzhanguan_20130301-20170228.csv"))
df6 = pd.read_csv(Path("Resources/PRSA_Data_Shunyi_20130301-20170228.csv"))
df7 = pd.read_csv(Path("Resources/PRSA_Data_Tiantan_20130301-20170228.csv"))
df8 = pd.read_csv(Path("Resources/PRSA_Data_Wanliu_20130301-20170228.csv"))
df9 = pd.read_csv(Path("Resources/PRSA_Data_Wanshouxigong_20130301-20170228.csv"))
df10 = pd.read_csv(Path("Resources/PRSA_Data_Aotizhongxin_20130301-20170228.csv"))
df11 = pd.read_csv(Path("Resources/PRSA_Data_Changping_20130301-20170228.csv"))
df12 = pd.read_csv(Path("Resources/PRSA_Data_Dingling_20130301-20170228.csv"))
                                                                     
# Review the DataFrame
display(df1.head())
display(df1.tail())
```

```python
df1.isna().sum()
```

```python
# make a list of the df's
list_df = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12]

# for loop to check number of NaN issues in all dataframes
na_info = {}

# for i in list_df:
#     i.isna().sum()

for i in list_df:
    na_info = i.isna().sum()
    print(na_info)

```

```python
# fillNaN with median value and write out df1_med
df1_med = df1.fillna(df1.median())
df1_med.isna().sum() # points out some Wind Direction Na's (not int) so median doesn't address - add dropna?
# df1_med.head()
```

```python
df1_med = df1_med.dropna()
df1_med.isna().sum()
```

```python
list_df_med = []

# for loop to fillNaN with median value and write out df1_med
for i in list_df:
    i2 = i.fillna(i.median())
    list_df_med.append(i2)
    
```

```python
df12_med = list_df_med[11]
df12_med.head()
```

```python
# dropNaN and write out df1_drop
df1_drop = df1.dropna()
df1_drop.isna().sum()
# display(df1_drop.count())  shows that we dropped 5000 rows or about 15% of the data
```

```python
# for loop to dropNaN and write out df1_drop
list_df_drop = []

for i in list_df:
    i2 = i.dropna()
    list_df_drop.append(i2)
```

```python
df11_drop = list_df_drop[10]
df11_drop.head()
```

```python


#concatonate the original data fames
data = pd.concat(list_df)

#concatonate the data frames where numeric NA's were filled with median values for each data frame
data_med = pd.concat(list_df_med)

#concatonate the data frames where numeric NA's were dropped 
data_drop = pd.concat(list_df_drop)

# Review the DataFrame
display(data.head())
display(data.tail())
display(data.count())  # to count total rows in dataframe
# display(data.count()) identified that approximately 5% of data for NO2, CO, and O3 is NaN prior to fillNA
print(f"the number of rows in df data is {len(data)} after merging the original csv files.")
print(f"the number of rows in df data_med is {len(data_med)} after filling numeric NA's with median values.")
print(f"the number of rows in df data_drop is {len(data_drop)} after dropping numeric NA's")  
# df data_drop shows about 10% of rows were dropped
```
